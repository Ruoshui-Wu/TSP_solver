```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xy = np.random.rand(100,2)
x,y = xy.transpose()
```

```python
# TSP_Drawer can draw the solution step by step, which is given by trained neural network. 
class TSP_Drawer:
    def __init__(self,xy):
        self.x,self.y = xy.transpose()
        self.fig, self.ax = plt.subplots(figsize=(10,10))
        self.ax.set_xlim(0,1)
        self.ax.set_ylim(0,1)

    def animation(self):
        self.fig.canvas.mpl_connect("button_press_event", self.mouse_event)
        self.ax.scatter(self.x, self.y, s=40, color='blue')
        plt.show()
    def mouse_event(self, event):
        self.x.append(int(event.xdata))
        self.y.append(int(event.ydata))
        
        
        
```

```python
drawer = TSP_Drawer(xy)
drawer.animation()
```
